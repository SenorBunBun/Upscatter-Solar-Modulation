# Generating Background Data

Importing Packages.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def genPoisson(mean, size = None):

    return np.random.poisson(mean, size)

def genGaussian (mean, size = None):

    return np.random.normal(mean, np.sqrt(mean), size)
```

```python
lambdaB = 1
num = 100
fig, axe = plt.subplots(4, 5, figsize=(25,15))
plt.setp(axe, xlabel='Events', ylabel='# of Occur')
fig.tight_layout()
for row in range(4):
    for col in range(5):
        axe[row, col].hist(genGaussian(lambdaB, 300), bins= 30, histtype = 'step')
        axe[row, col].hist(genPoisson(lambdaB, 300), bins= 30, histtype = 'step')
        axe[row, col].set_title('Mean of {0}'.format(lambdaB))
        lambdaB += 1

plt.show()


```
