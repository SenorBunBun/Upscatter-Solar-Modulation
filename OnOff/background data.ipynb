# Generating Background Data

Importing Packages.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def genPoisson(N, T, Tau):
    mean = (N * T)/Tau
    return np.random.poisson(mean)

def genGaussian(N, T, Tau):
    mean = (N * T)/Tau
    return np.random.normal(mean, np.sqrt(mean))
```

```python
Num = 0.01
decayRate = 0.77 #Half life of B^8 in seconds
timeBin = 18000 # 5 hours in seconds
mintimeBin = 5
timeStep = (timeBin - mintimeBin)/300

tdata = []
poissondata = []
gaussiandata = []
```

```python
while timeBin > mintimeBin: # 20 minute time scale
    gaussiandata.append(genGaussian(Num, timeBin, decayRate))
    poissondata.append(genPoisson(Num, timeBin, decayRate))
    tdata.append(timeBin)
    timeBin += -timeStep
```

```python
rangeP = (min(poissondata), max(poissondata))
rangeT = (min(tdata), max(tdata))
```

```python
plt.yticks(  np.arange(rangeP[0], rangeP[1], np.round((rangeP[1] - rangeP[0])/15) )  )


plt.xticks( np.arange(rangeT[0], rangeT[1], np.round((rangeT[1] - rangeT[0])/5) )  )


plt.plot(tdata, poissondata, label = "Poisson Random Vals")
plt.plot(tdata, gaussiandata, label = "Gaussian Random Vals")

plt.legend()
plt.show()


```
