# Generating Background Data

Importing Packages.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def genPoisson(N, T, Tau):
    mean = (N * T)/Tau
    return np.random.poisson(mean)

def genGaussian(N, T, Tau):
    mean = (N * T)/Tau
    return np.random.normal(mean, np.sqrt(mean))
```

```python
Num = 1
decayRate = 50 #Half life of B^8 in seconds
timeBin = 12 # 5 hours in seconds
mintimeBin = 0.0166666667
timeStep = (timeBin - mintimeBin)/300

tdata = []
poissondatabin = []
gaussiandatabin = []
```

```python
while timeBin > mintimeBin: # 20 minute time scale
    gaussiandatabin.append(genGaussian(Num, timeBin, 1/decayRate))
    poissondatabin.append(genPoisson(Num, timeBin, 1/decayRate))
    tdata.append(timeBin)
    timeBin += -timeStep
```

```python
rangeP = (min(poissondatabin), max(poissondatabin))
rangeT = (min(tdata), max(tdata))
```

```python
plt.yticks(  np.arange(rangeP[0], rangeP[1], np.round((rangeP[1] - rangeP[0])/15) )  )

plt.xticks( np.arange(rangeT[0], rangeT[1], np.round((rangeT[1] - rangeT[0])/5) )  )


plt.plot(tdata, poissondatabin, label = "Poisson Random Vals")
plt.plot(tdata, gaussiandatabin, label = "Gaussian Random Vals")

plt.legend()
plt.show()
```
