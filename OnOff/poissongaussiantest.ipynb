# Generating Background Data

Importing Packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as kfp
import scipy.integrate as integrate
```

```python
def genPoisson(mean, size = 1):

    return np.random.poisson(mean, size)

def genGaussian (mean, size = 1):
    num = np.random.normal(mean, np.sqrt(mean),size)
    return list(map(lambda x: 0 if(x<0) else x, list(np.rint(num))))
```

```python
lambdaB = 1
num = 300

fig, axe = plt.subplots(4, 5, figsize=(25,15))
plt.setp(axe, xlabel='Number of Events', ylabel='Number of Occurrences in A Bin')
fig.tight_layout()

for row in range(4):
    for col in range(5):
        axe[row, col].hist(genGaussian(lambdaB, int(np.round(np.multiply(num, np.sqrt(lambdaB) ) ) )), label='Gaussian', bins= 30, histtype = 'step')
        axe[row, col].hist(genPoisson(lambdaB,  int(np.round(np.multiply(num, np.sqrt(lambdaB) ) ) )), label='Poisson', bins= 30, histtype = 'step')
        axe[row, col].set_title('Mean of {0}'.format(lambdaB))
        lambdaB += 1

fig.legend(['Gaussian', 'Poisson'])
fig.subplots_adjust(hspace=0.3)
plt.show()
```
